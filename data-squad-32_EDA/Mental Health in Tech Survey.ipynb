#### Filtering to get a DF with just the people from the USA, as that will be my main target 

```python
ds_usa = ds[ds['Country'] == 'United States']
ds_usa
```

```python
# Checking null count
ds_usa.isna().sum()
```

```python
# Normalizing the column names
ds_usa.columns = map(str.title, ds_usa.columns)
```

#### Creating a new column with the number of people that have or don't have a mental illness

```python
# Creating a new column with the number of people that have or don't have a mental illness
#ds_usa['Work_Interfere'] = ds_usa['Work_Interfere'].str.strip()

# Was trying to do it without loc, but couldn't figure out how
ds_usa.loc[ds_usa['Work_Interfere'] != 'NA', "Mentally_Ill"] = 'Yes'

# Filling the NaN's with 0, as they are the rest of the pop that didn't answer this question, and
# therefore they do not have a mental illness
ds_usa['Mentally_Ill'].fillna('No', inplace = True)

# Changing the type to int - was having doubts whether or not to do 0-1 or just leave as yes-no
#ds_usa['Mentally_Ill'] = ds_usa['Mentally_Ill'].astype(int)

# Checking the total we have here, and it now matches the total number of observations in the data
ds_usa['Mentally_Ill'].count() == len(ds_usa)
```

#### Cleaning up the 'Age' column

```python
'''
Removing some of the '-' signs I saw after doing describe, as someone had -29 years of age.
This way I'll get only the absolute values and the problem is solved. However, after doing this, I
noticed that some people still had wrong age values, as the min was 1 and max 329...
So I decided to just check for these values and remove them in the end, as I do below.
'''
ds_usa['Age'] = ds_usa['Age'].abs()

# Ok, I'm just gonna drop these 4 rows, since I don't know their correct age (329, 5, 11 and 1)
# Noted that index 734 had a nice comment, though
# Losing 3 ill and 1 not
ds_usa[(ds_usa['Age'] < 18) | (ds_usa['Age'] > 75)]

# Seems like I'm losing another 3 besides the 4 that had wrong age values. Don't know why
ds_usa = ds_usa[(ds_usa['Age'] > 18) & (ds_usa['Age'] < 75)]
#ds_usa['Age'] = ds_usa['Age'].astype(str)
#ds_usa['Age'] = ds_usa['Age'].str.strip()
#ds_usa['Age'] = ds_usa['Age'].astype(int)
#ds_usa.dtypes

# Now the ages seem to be corrected without the 7 values I dropped. Total is 744
#ds_usa['Age'].value_counts()
```

#### Cleaning up the 'Gender' column

```python
# Making all the values be lowercase
ds_usa['Gender'] = ds_usa['Gender'].str.lower()

# Removing the spaces
ds_usa['Gender'] = ds_usa['Gender'].str.strip()

# Replacing the values with male/female
ds_usa['Gender'].replace({'femail': 'female', 'f': 'female', 'cis female': 'female', 
                          'femake': 'female', 'woman': 'female', 'female (cis)': 'female', 
                          'cis-female/femme': 'female', 'm': 'male', 'male-ish': 'male', 
                          'maile': 'male', 'male (cis)': 'male', 'cis male': 'male', 
                          'make': 'male', 'man': 'male', 'msle': 'male', 
                          'mail': 'male', 'female (trans)': 'female', 
                          'trans woman': 'female', 'trans-female': 'female', 
                          'nah': 'other', 'genderqueer': 'other', 
                          'non-binary': 'other', 'queer/she/they': 'other'}, inplace = True)

# Checking if the cleaning worked
print('Uniques:', ds_usa['Gender'].unique())
print('')
print('Value count:', ds_usa['Gender'].value_counts(), sep = '\n')
```

## STEP 2 - Analysis

### Illustrating the amount of mentally ill people in the USA vs the ones that sought treatment for mental health issues

```python
fig, ax = plt.subplots(1, 2, figsize = (20, 10))
#plt.subplots_adjust(wspace = 0.3) # after I increased the size this wasn't necessary anymore :(

# PIE CHART ILLUSTRATING THE % OF PEOPLE THAT HAVE A MENTAL ILLNESS IN THE USA
mental = ds_usa['Mentally_Ill'].value_counts()
mental.plot(kind = 'pie', legend = 1, autopct='%1.1f%%', colors = ['darkgreen', 'darkorange'],
            title = 'Percentage of mentally ill people over total US sample', fontsize = 20, 
            ax = ax[0], label = 'Mentally Ill');

# PIE CHART ILLUSTRATING THE % OF PEOPLE THAT SOUGHT PSYCHOLOGICAL TREATMENT IN THE USA
treatment = ds_usa['Treatment'].value_counts()
treatment.plot(kind = 'pie', legend = 1, autopct='%1.1f%%', colors = ['darkgreen', 'darkorange'], 
            title = 'Percentage of people that sought psychological treatment over total US sample', 
                      fontsize = 20, startangle = 90, ax = ax[1]);

plt.savefig('Mentally-ill-and-treatment')
```

### Illustrating the total gender percentage in the USA vs the gender count regarding mental health issues or not

```python
#fig, ax = plt.subplots(1, 2, figsize = (20, 10))

# PIE CHART ILLUSTRATING THE % OF THE GENDER IN THE USA SAMPLE
gender = ds_usa['Gender'].value_counts()
gender.plot(kind = 'pie', legend = 1, autopct='%1.1f%%', colors = ['darkgreen', 'darkorange', 'blue'],
            title = 'Percentage of genders over total US sample', fontsize = 20);#, 
            #ax = ax[0], label = 'Gender');

plt.savefig('Gender-percentage')

# BAR CHART ILLUSTRATING THE AMOUNT MALE/FEMALE/OTHER THAT HAVE OR DON'T HAVE A MENTAL ILLNESS
plt.figure(figsize = (12, 6))
sns.countplot(x = 'Gender', hue = 'Mentally_Ill', data = ds_usa );
#ax[1].plot(mental);
#ax[1].plot(treatment);

plt.savefig('Gender-illness-count')
```
